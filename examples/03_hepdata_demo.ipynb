```python
import numpy as np
import matplotlib.pyplot as plt
from hepviz.io import load_root_file

# Load the demo ROOT file
tree = load_root_file("../data/sample_small.root", treename="Events")

print("Available branches:", list(tree.keys()))
```

```python
px = tree["px"].array(library="np")
py = tree["py"].array(library="np")
pz = tree["pz"].array(library="np")
E  = tree["E"].array(library="np")

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0,0].hist(px, bins=50, histtype='step', color='blue'); axs[0,0].set_title("px distribution")
axs[0,1].hist(py, bins=50, histtype='step', color='red'); axs[0,1].set_title("py distribution")
axs[1,0].hist(pz, bins=50, histtype='step', color='green'); axs[1,0].set_title("pz distribution")
axs[1,1].hist(E,  bins=50, histtype='step', color='black'); axs[1,1].set_title("Energy distribution")
plt.tight_layout()
plt.show()
```

```python
mass = np.sqrt(np.maximum(E**2 - (px**2 + py**2 + pz**2), 0))

plt.figure(figsize=(6,4))
plt.hist(mass, bins=60, histtype='step', color='purple')
plt.xlabel("Invariant Mass [GeV]")
plt.ylabel("Events")
plt.title("Invariant Mass Distribution")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

ax.quiver(np.zeros(200), np.zeros(200), np.zeros(200),
          px[:200], py[:200], pz[:200], length=0.05, normalize=True, color='blue')

ax.set_xlabel('px')
ax.set_ylabel('py')
ax.set_zlabel('pz')
ax.set_title("3D Momentum Vectors (first 200 events)")
plt.show()
```
