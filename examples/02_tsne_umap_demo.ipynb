```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.manifold import TSNE
import umap.umap_ as umap
from mpl_toolkits.mplot3d import Axes3D  # needed for 3D projection

# --- Generate synthetic dataset ---
X, y = make_classification(
    n_samples=500,
    n_features=50,
    n_informative=10,
    n_classes=3,
    random_state=42
)

# --- 2D embeddings ---
X_tsne_2d = TSNE(n_components=2, random_state=42).fit_transform(X)
X_umap_2d = umap.UMAP(n_components=2, random_state=42).fit_transform(X)

# --- 3D embeddings ---
X_tsne_3d = TSNE(n_components=3, random_state=42).fit_transform(X)
X_umap_3d = umap.UMAP(n_components=3, random_state=42).fit_transform(X)

# --- Plot 2D results ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(X_tsne_2d[:, 0], X_tsne_2d[:, 1], c=y, cmap="tab10", s=20, alpha=0.7)
axes[0].set_title("t-SNE (2D)")

axes[1].scatter(X_umap_2d[:, 0], X_umap_2d[:, 1], c=y, cmap="tab10", s=20, alpha=0.7)
axes[1].set_title("UMAP (2D)")

plt.show()

# --- Plot 3D results ---
fig = plt.figure(figsize=(12, 5))

# t-SNE 3D
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X_tsne_3d[:, 0], X_tsne_3d[:, 1], X_tsne_3d[:, 2], c=y, cmap="tab10", s=20, alpha=0.7)
ax1.set_title("t-SNE (3D)")

# UMAP 3D
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(X_umap_3d[:, 0], X_umap_3d[:, 1], X_umap_3d[:, 2], c=y, cmap="tab10", s=20, alpha=0.7)
ax2.set_title("UMAP (3D)")

plt.show()
```
