```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=500, n_features=10, n_classes=2, random_state=42)
```

```python
# PCA 2D
X_pca2 = PCA(n_components=2).fit_transform(X)
plt.scatter(X_pca2[:,0], X_pca2[:,1], c=y, cmap="coolwarm", alpha=0.7)
plt.title("PCA 2D")
plt.show()
```

```python
# PCA 3D
from mpl_toolkits.mplot3d import Axes3D
X_pca3 = PCA(n_components=3).fit_transform(X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca3[:,0], X_pca3[:,1], X_pca3[:,2], c=y, cmap="coolwarm", alpha=0.7)
plt.title("PCA 3D")
plt.show()
```
